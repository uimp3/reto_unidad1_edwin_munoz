## Tratamiento de negativos y nulos, y normalización de fechas

```python
# Recargar el archivo ignorando comillas separadoras
df = pd.read_csv(
    "../data/ventas_techventas.csv",
    quoting=csv.QUOTE_NONE,
    encoding="latin1"
)

df.head()
```

## Detección de nulos y negativos de la nueva carga

```python
df[
    df.isna().any(axis=1) |
    (df["cantidad"] < 0)
]
```

```python
print("tamaño del dataset original", df.size)
''' Para producto nulo en el registro 42 se realiza:
Rellenar el nulo basado en la coincidencia de precio y categoría '''
df.loc[(df['producto'].isna()) & (df['precio_unitario'] == 350.0), 'producto'] = 'Monitor 27'

# Para cantidades negativas, se elimina el registro
df = df[df["cantidad"] > 0]

# Normalización de fechas
df["fecha"] = pd.to_datetime(
    df["fecha"],
    errors="coerce"
)
df = df[df["fecha"].notna()]
print("tamaño del dataset limpio", df.size)
```

# 1.2 Álgebra Lineal Aplicada
## Arrays

```python
cantidades = df["cantidad"].to_numpy()
precios = df["precio_unitario"].to_numpy()
```

## Cálculo de Ingresos Brutos

```python
# con operador @
ingresos = cantidades @ precios
print("ingresos totales: ",ingresos)

ingresos_por_venta = cantidades * precios
print(ingresos_por_venta)
```

## 1.3 Estadísticas con NumPy

```python
precios = df["precio_unitario"].dropna().to_numpy()

print("Media:", np.mean(precios))
print("Mediana:", np.median(precios))
print("Std:", np.std(precios))
print("P25:", np.percentile(precios,25))
print("P75:", np.percentile(precios,75))
```

## Interpretación de las estadísticas descriptivas

Se calcularon la media, mediana, desviación estándar y percentiles del campo **precio_unitario** utilizando NumPy.

### Resultados obtenidos

| Métrica | Valor |
|----------|----------:|
| Media | 364.32 |
| Mediana | 120.00 |
| Desviación estándar | 435.66 |
| Percentil 25 | 90.00 |
| Percentil 75 | 350.00 |

### Análisis

La **media** de los precios es de **364.32**, mientras que la **mediana** es de **120.00**. La diferencia significativa entre ambos valores indica que la distribución de precios no es simétrica y se encuentra sesgada hacia la derecha. Esto sugiere la existencia de productos de alto valor que elevan considerablemente el promedio general del catálogo.

La **desviación estándar** de **435.66** es incluso superior a la mediana, lo que evidencia una alta dispersión en los precios. En otras palabras, el catálogo contiene productos con rangos de precio muy diferentes, desde artículos relativamente económicos hasta productos considerablemente más costosos.

El **percentil 25 (P25)** indica que el 25% de los productos tienen precios iguales o inferiores a **90.00**, mientras que el **percentil 75 (P75)** muestra que el 75% de los productos tienen precios iguales o inferiores a **350.00**. Esto significa que el 50% central de los precios se concentra entre **90.00 y 350.00**.

### Conclusiones sobre el catálogo

Los resultados sugieren que TechVentas S.A.S. maneja un catálogo diversificado con productos de distintos segmentos de precio. Aunque la mayoría de los artículos se concentra en rangos moderados (entre 90 y 350 unidades monetarias), existen algunos productos de alto valor que incrementan significativamente el promedio general.

Esta situación es común en empresas tecnológicas donde conviven accesorios de bajo costo, como periféricos o componentes básicos, con equipos de mayor valor, como computadores, servidores o dispositivos especializados. Por ello, la mediana representa mejor el precio típico del catálogo que la media, ya que es menos sensible a valores extremos.
